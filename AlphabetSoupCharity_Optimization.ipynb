## Preprocessing the Data

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf
import os

#  Import and read the charity_data.csv.
import pandas as pd 
df = pd.read_csv("../Resources/charity_data.csv")
df.head()
```

```python
df = df.drop("EIN", 1)
df = df.drop("NAME", 1)
df
```

```python
df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
#df['APPLICATION_TYPE'].value_counts()
application_counts = df['APPLICATION_TYPE'].value_counts()
application_counts
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_types_to_replace = list(application_counts[application_counts < 500].index)

# Replace in dataframe
for app in application_types_to_replace:
    df['APPLICATION_TYPE'] = df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
# df['CLASSIFICATION'].value_counts()
class_counts = df['CLASSIFICATION'].value_counts()
class_counts
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
class_count_1=class_counts[class_counts>1]
class_count_1
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_replace = list(class_counts[class_counts < 1000].index)

# Replace in dataframe
for cls in classifications_to_replace:
    df['CLASSIFICATION'] = df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
dummies_df = pd.get_dummies(df, drop_first = True)
dummies_df
```

```python
# Split our preprocessed data into our features and target arrays
X = dummies_df.drop(["IS_SUCCESSFUL"], axis='columns')#.values
y = dummies_df["IS_SUCCESSFUL"]#.values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
inputs = len(X.columns)
```

```python
print(inputs)
```

## Compile, Train and Evaluate the Model - Second Attempt

```python
# Create a method that creates a new Sequential model with hyperparameter options
def create_model(hp):
    nn_model = tf.keras.models.Sequential()

    # Allow kerastuner to decide which activation function to use in hidden layers
    activation = hp.Choice('activation',['relu','tanh'])
    
    # Allow kerastuner to decide number of neurons in first layer
    nn_model.add(tf.keras.layers.Dense(units=hp.Int('first_units',
        min_value=6,
        max_value=18,
        step=6), activation=activation, input_dim=inputs))

    # Allow kerastuner to decide number of hidden layers and neurons in hidden layers
    for i in range(hp.Int('num_layers', 1, 2)):
        nn_model.add(tf.keras.layers.Dense(units=hp.Int('units_' + str(i),
            min_value=6,
            max_value=12,
            step=6),
            activation=activation))
    
    nn_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

    # Compile the model
    nn_model.compile(loss="binary_crossentropy", optimizer='adam', metrics=["accuracy"])
    
    return nn_model
```

```python
# Import the kerastuner library
import keras_tuner as kt

tuner = kt.Hyperband(
    create_model,
    objective="val_accuracy",
    max_epochs=20,
    hyperband_iterations=2,
directory=os.path.normpath('C:/')
)
```

```python
# Run the kerastuner search for best hyperparameters
tuner.search(X_train_scaled,y_train,epochs=20,validation_data=(X_test_scaled,y_test))
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
#  YOUR CODE GOES HERE

nn_model = tf.keras.models.Sequential()

# First hidden layer
nn_model.add(tf.keras.layers.Dense(units=18, activation="relu", input_dim=36))

# Second hidden layer
nn_model.add(tf.keras.layers.Dense(units=18, activation="sigmoid"))

# Output layer
nn_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn_model.summary()
```

```python
# Compile the model
nn_model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
from keras.callbacks import Callback

mc = tf.keras.callbacks.ModelCheckpoint('weights{epoch:08d}.h5', 
                                     save_weights_only=True, period=5)
    
        
fit_model = nn_model.fit(X_train_scaled, y_train, epochs=30, callbacks = [mc])
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn_model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

## Analysis Overview

The purpose of this analysis is to determine an algorithm which will predict success for the nonprofit organizations applying to Alphabet Soup Charity for funding. 

## Results:

## Data Preprocessing:
- The target variable for the model is if the charities are effectively using their money received from Alphabet Soup, and this is seen in the "IS_SUCCESSFUL" column.
- The variables considered to be the features of the model are the type of application, (e.g T2, T9, T19, etc.) and classification according to type of government organization (e.g. C1000, C2100, C4500, etc.)
- The EIN and names of the organizations are not considered the target or features of the model, and therefore was dropped from the data frame. 

## Compiling, Training, and Evaluating the Model:
- For the initial model, 18 neurons were chosen for each of the first and second layers due to the dimensionality of the data (this being twice the number of the binned application types and three times the number of classification types.)
- Two hidden layers were chosen for each of the features of the model. Relu was chosen as the activation function for the initial two hidden layers. Since Relu does not need to perform exponential operations, use of this activation model is quicker and more reliable than either sigmoid or tanh. Sigmoid was used in the output layer to draw a distinction between two alternatives (whether or not the charity will be successful in its use of funding).
- Over four attempts to optimize the model, I was unable to meet the 75% target predictive accuracy. 
- During the second attempt, I used the Keras Tuner model to let the computer pick the best number of neurons and layers as well as activation functions. The Keras Tuner model yielded an accuracy, when rounded, is identical to the initial accuracy of 0.728. 
- In the third attempt, I modified the activation model of the layer to sigmoid, and got a slightly reduced accuracy of 0.725. 
- For the final attempt, I decided to further pursue the activation models and designate the first layer as relu, but the second layer as tanh, and the third layer remained sigmoid. I left the number of neurons the same so that no other variables would explain the accuracy obtained. The fourth and final accuracy obtained was 0.727, slightly better than the third attempt but still just under the initial and second model accuracy of 0.728. 

## Summary:

- The best results for the models were yielded with each hidden layer being a relu function, and allowing the Keras Tuner to select neurons, layers and activation functions. 
- For further models, I would add more neurons to each layer, more hidden layers based on the number of features, and vary the number of epochs over which the model is trained. 
- Additional features could include further binning of the application types and classification numbers. In addition, the amount of times a charity has received funding could be used as an additional feature. 
- A recommendation for another model to use in this situation would be the Random Forest model. This model is good for classification problems because there is a robust number of decision trees. It is also not affected by overfitting.
